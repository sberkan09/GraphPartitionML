```python
import networkx as nx
import random
import matplotlib.pyplot as plt

def create_graph_examples(num_graphs_per_type):
    graph_examples = []

    # Tree
    for _ in range(num_graphs_per_type):
        n = random.randint(5, 10)
        graph_examples.append(nx.random_tree(n=n))

    # Simple Graph
    for _ in range(num_graphs_per_type):
        n = random.randint(5, 10)
        p = random.uniform(0.1, 0.5)
        graph_examples.append(nx.gnp_random_graph(n=n, p=p))

    # Complete Graph
    for _ in range(num_graphs_per_type):
        n = random.randint(3, 10)
        graph_examples.append(nx.complete_graph(n=n))

    # Null Graph
    for _ in range(num_graphs_per_type):
      graph_examples.append(nx.null_graph())


    # Trivial Graph
    for _ in range(num_graphs_per_type):
        graph_examples.append(nx.trivial_graph())

    # Multi Graph
    for _ in range(num_graphs_per_type):
        n = random.randint(3, 10)
        m = nx.MultiGraph()
        m.add_nodes_from(range(n))
        # Add some random edges
        for i in range(random.randint(1, n*(n-1)//2)):
            m.add_edge(random.randint(0, n-1), random.randint(0, n-1))
        graph_examples.append(m)

    # Pseudo Graph
    for _ in range(num_graphs_per_type):
        n = random.randint(3, 10)
        m = nx.MultiGraph()
        m.add_nodes_from(range(n))
        # Add some edges including self loops (pseudo edges)
        for i in range(random.randint(1, n*(n-1)//2)):
            m.add_edge(random.randint(0, n-1), random.randint(0, n-1))
        # Add one self-loop
        m.add_edge(random.randint(0, n-1), random.randint(0, n-1))
        graph_examples.append(m)

    # Cyclic Graph
    for _ in range(num_graphs_per_type):
        n = random.randint(3, 10)
        graph_examples.append(nx.cycle_graph(n=n))

    # Bipartite Graph
    for _ in range(num_graphs_per_type):
        n1 = random.randint(2, 5)
        n2 = random.randint(2, 5)
        graph_examples.append(nx.complete_bipartite_graph(n1, n2))

    # Euler Graph
    for _ in range(num_graphs_per_type):
        # Create a graph with an Eulerian circuit: even number of nodes and random edges
        n = random.randint(5, 10) // 2 * 2  # even number of nodes
        euler_graph = nx.MultiGraph()
        euler_graph.add_nodes_from(range(n))
        # Add random edges ensuring each node has an even degree
        for i in range(n):
            for _ in range(n//2):
                euler_graph.add_edge(i, (i + random.randint(1, n-1)) % n)
        graph_examples.append(euler_graph)

    # Hamiltonian Graph
    for _ in range(num_graphs_per_type):
        n = random.randint(5, 10)
        hamiltonian_graph = nx.cycle_graph(n)
        # Add some random edges but keep the graph Hamiltonian
        extra_edges = random.randint(0, n//2)
        for _ in range(extra_edges):
            source = random.randint(0, n-1)
            target = (source + random.randint(2, n-1)) % n
            hamiltonian_graph.add_edge(source, target)
        graph_examples.append(hamiltonian_graph)

    # Visualization (Optional)
    for graph in graph_examples:
        nx.draw(graph, with_labels=True)
        plt.show()

    return graph_examples

# Example usage:
graphs = create_graph_examples(2)  # This will create 2 examples of each graph type
```
